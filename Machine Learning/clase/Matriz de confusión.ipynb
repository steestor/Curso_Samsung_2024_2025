```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Cargar los datos de Iris
iris = load_iris()
X = iris.data
y = iris.target

# Convertir el problema a clasificación binaria (usamos solo 2 clases)
# Aquí seleccionamos solo las dos primeras clases para simplificar
X = X[y != 2]  # Eliminar la clase 2
y = y[y != 2]  # Eliminar la clase 2

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de clasificación
clf = RandomForestClassifier()

# Entrenar el modelo
clf.fit(X_train, y_train)

# Predecir las clases en el conjunto de prueba
y_pred = clf.predict(X_test)
```

```python
# Calcular la matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Imprimir la matriz de confusión
print("Matriz de Confusión:")
print(cm)
```

```python
# Calcular e imprimir el informe de clasificación (precisión, recall, F1-score)
print("Informe de clasificación:")
print(classification_report(y_test, y_pred))
```
