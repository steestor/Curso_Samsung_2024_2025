# Descripción de las Variables del Dataset de Viviendas de Boston

* **TasaCriminalidad**: Tasa de criminalidad per cápita por ciudad.
* **PorcentajeLotesGrandes**: Porcentaje de lotes residenciales de más de 25,000 pies cuadrados.
* **PorcentajeTerrenoIndustrial**: Proporción de acres de negocios no minoristas por ciudad.
* **CercaRioCharles**: Variable indicadora del río Charles (1 si el tramo limita con el río; 0 en caso contrario).
* **OxidosNitrogeno**: Concentración de óxidos de nitrógeno (partes por 10 millones).
* **PromedioHabitacionesVivienda**: Número promedio de habitaciones por vivienda.
* **PorcentajeEdificiosAnteriores1940**: Proporción de unidades ocupadas por sus propietarios construidas antes de 1940.
* **DistanciaCentrosEmpleo**: Distancias ponderadas a cinco centros de empleo de Boston.
* **AccesibilidadAutopistas**: Índice de accesibilidad a autopistas radiales.
* **TasaImpuestoPropiedad**: Tasa de impuesto a la propiedad de valor total por $10,000.
* **ProporcionAlumnosMaestros**: Proporción de alumnos por maestro por ciudad.
* **PorcentajePoblacionAfroamericana**: 1000(Bk - 0.63)^2 donde Bk es la proporción de residentes afroamericanos por ciudad.
* **PorcentajePoblacionEstatusInferior**: Porcentaje de la población de estatus socioeconómico inferior.
* **ValorMedianoViviendas**: Valor mediano de las viviendas ocupadas por sus propietarios en $1000.

```python
columnas = [
    'TasaCriminalidad',
    'PorcentajeLotesGrandes',
    'PorcentajeTerrenoIndustrial',
    'CercaRioCharles',
    'OxidosNitrogeno',
    'PromedioHabitacionesVivienda',
    'PorcentajeEdificiosAnteriores1940',
    'DistanciaCentrosEmpleo',
    'AccesibilidadAutopistas',
    'TasaImpuestoPropiedad',
    'ProporcionAlumnosMaestros',
    'PorcentajePoblacionAfroamericana',
    'PorcentajePoblacionEstatusInferior',
    'ValorMedianoViviendas'
]
data.columns = columnas
data.head()
```
