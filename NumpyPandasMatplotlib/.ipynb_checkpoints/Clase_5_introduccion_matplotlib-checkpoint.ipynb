**Matplotlib** es una biblioteca de Python que permiten crear gran variedad de gráficas a partir de datos almacenados en arrays o listas. Matplotlib contiene, entre otras cosas, un módulo llamado **pyplot** que proporciona las funciones necesarias para crear las gráficas.

> **ATENCIÓN:**
>
> Es importante tener en cuenta que este módulo proporciona dos interfaces alternativas (por tanto diferentes) para crear las gráficas:
>
> 1. Interfaz funcional (también llamado interfaz `pyplot`).
> 2. Intefaz Orientado a Objetos (OO).
>
> En este tutorial vamos a centrarnos en la intefaz **orientada a objetos.**

Para más información puedes ir a la siguiente página https://matplotlib.org/stable/tutorials/introductory/lifecycle.html

Para empezar a usarlo, lo primero que debemos hacer es importarlo:

````python
import matplotlib.pyplot as plt
````

## Una gráfica simple

Un modo sencillo de craer una gráfica a partir de datos almacenados en dos listas.

- Lo primero es crear un objeto de tipo Figure y un objeto de tipo Axes:

    ```
    fig, ax = plt.subplots()
    ```
    
- El objeto `ax` representa una gráfica individual, más adelante veremos que es posible crear varias gráficas en una misma figura.
- Sobre `ax` aplicamos el método `plot`.

````python
x = [1,2,3, 4, 5] # Valores para el eje x
y = [1,4,9,16,25] # Valores de y para cada x

# lo primero es crear un objeto de tipo Figure y un objeto de tipo Axes:
fig, ax = plt.subplots()
# utilizamos ax para dibujar la gráfica
ax.plot(x,y)
None # para evitar que salga una línea tipo: [<matplotlib.lines.Line2D at 0x7f7efd139900>]
````

Si no se proporcionan valores para el eje x, por defecto se toman valores enteros consecutivos empezando en 1. El código anterior sería equivalente a:

````python
y = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(y) # Se toma por defecto valores 1,2,3...etc para el eje x
None # para evitar que salga una línea tipo: [<matplotlib.lines.Line2D at 0x7f7efd139900>]
````

### Lo mismo, con datos almacenados en arrays numpy
Los datos también pueden estar almacenados en arrays numpy. De hecho, si los datos se pasan en forma de lista, matplotlib los convierte internamente a un array numpy. El siguente código generará exactamente la misma gráfica

````python
import numpy as np
y = np.array([1,4,9,16,25]) # Valores almacenados en un array numpy
fig, ax = plt.subplots()
ax.plot(y)                 # Crear la gráfica
None
````

### Personalización: etiquetas, títulos y rejilla
Es posible añadir etiquetas a cada uno de los ejes y un título a la gráfica. También se puede sobreimprimir una rejilla

````python
x = [1,2,3, 4, 5] # Valores para el eje x
y = [1,4,9,16,25] # Valores de y para cada x
fig, ax = plt.subplots()
ax.plot(y)                    # Crear la gráfica
ax.set_title("x elevado a 2") # Título
ax.set_xlabel("x")            # Etiqueta para el eje x
ax.set_ylabel("$x^2$")        # Etiqueta para el eje y
ax.grid()                     # Sobreimprimir rejilla
None
````

Además de los métodos `set_title`, `set_xlabel`, `set_ylabel`, etc. es posible pasar esta información mediante un método `set` descrito en https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set.html

El ejemplo anterior utilizando `set` quedaría así:

````python
x = [1,2,3, 4, 5] # Valores para el eje x
y = [1,4,9,16,25] # Valores de y para cada x
fig, ax = plt.subplots()
ax.plot(y)
ax.set(title="x elevado a 2",
       xlabel="x",
       ylabel="$x^2$")
ax.grid()
None
````

## Varias funciones en una misma gráfica: leyendas
Se puede, en la misma gráfica, representar más de una función. En ese caso, además, conviene asignar una leyenda a cada función representada.

````python
x = [1,2,3,4,5]              # Valores para el eje x
y = [1,4,9,16,25]            # Unos valores en función de x
z = [1,8,27,64,125]          # Otros valores en función de x
fig, ax = plt.subplots()
ax.plot(x,y)                # Representar los valores de y
ax.plot(x,z)                # Representar los valores de z
ax.set_title("$x^2$ vs $x^3$")  # Título
ax.set_xlabel("x")              # Etiqueta para el eje x
ax.set_ylabel("y")              # Etiqueta para el eje y
ax.legend(["$X^2$","$x^3$"])# Leyendas. Se proporciona una lista con una
                             # leyenda para cada una de las funciones representadas
None
````

También es posible especificar la leyenda en cada llamada a `ax.plot` en lugar de utilizar `ax.legend(["$x^2$", "$x^3$"])`, aunque sigue siendo necesario llamar a `ax.legend()` como se muestra a continuación:

````python
x = [1,2,3,4,5]                # Valores para el eje x
y = [1,4,9,16,25]              # Unos valores en función de x
z = [1,8,27,64,125]            # Otros valores en función de x
fig, ax = plt.subplots()
ax.plot(x,y,label="$x^2$")     # Representar los valores de y
ax.plot(x,z,label="$x^3$")     # Representar los valores de z
ax.set(title="$x^2$ vs $x^3$",
       xlabel="x", ylabel="y")
ax.legend()                    # Hace falta de todos modos

None
````

## Personalización: tipo de línea y color

Los métodos para dibujar en un objeto de tipo `Axes` admiten argumentos que nos permiten 
personalizar las gráficas eligiendo el color, tipo de línea, etc.

Existen al menos dos maneras de proporcionar esta información:

1. Una forma abreviada mediante un argumento posicional que es una cadena donde se indica el color, tipo de línea, etc. Esta cadena está formada por la concatenación de un color (p. ej `'b'`, `'g'`, `'r'` para blue, green y red respectivamente) y un estilo de línea (p.ej. `'o'`, `'+'`, `'^'`, `'-'`, `'--'` para punto, símbolo +, triángulo, línea contínua y línea discontínua respectivamente). Por ejemplo, la cadena `'bo'` representa los datos mediante puntos azules.

2. Utilizando argumentos opcionales con nombre. Por ejemplo, `plot` (y otros métodos similares) admite:

    - `color` recibe una cadena de texto con el nombre del color en inglés (ejemplo: `"green"` o `"red"`).
    - `marker` recibe una cadena con el tipo de marcador. La lista se puede consultar en este enlace: https://matplotlib.org/stable/api/markers_api.html pero los más habituales son:
        - `"."` point
        - `"o"` circle
        - `"*"` star
        - `"+"` plus
    - `linestyle` permite seleccionar el estilo de la línea. Los más habituales son `"None"` (sin línea), `"solid"`, `"dotted"`, `"dashed"` y `"dashdot"`.


A continuación se muestran un ejemplo con la forma abreviada y a continuación el mismo ejemplo con los argumentos tipo keywords:

````python
# forma abreviada
x = [1,2,3,4,5]
y = [1,4,9,16,25]   
z = [1,8,27,64,125]
fig, ax = plt.subplots()
ax.plot(x, 'r--')
ax.plot(y, 'g+-.')
ax.plot(z, 'bo')
ax.legend(['x', '$x^2$', '$x^3$'])
None
````

````python
x = [1,2,3,4,5]
y = [1,4,9,16,25]   
z = [1,8,27,64,125]
fig, ax = plt.subplots()
ax.plot(x, color='red', linestyle='dashed')
ax.plot(y, color='green', marker='+', linestyle='dashdot')
ax.plot(z, color='blue', marker='o', linestyle='None')
ax.legend(['x', '$x^2$', '$x^3$'])
None
````

## Calcular los datos a visualizar automáticamente como *y=f(x)* 

En ocasiones los datos a mostrar deben calcularlos el propio programa mediante alguna función. En este caso, lo habitual es:

1. crear un array con valores consecutivos de x en el rango deseado y, a continuación,
2. calcular los valores de y para cada x.

Para (1) se puede usar la función `linspace` disponible en `numpy` en la que los dos primeros argumentos indican el rango de la variable x (mínimo y máximo) y el tercero la cantidad de valores (equidistantes) que se quiere en ese rango. Por ejemplo, `np.linspace(1,10,101)` devuelve un array de 101 elementos con los valores 1, 1.1, 1.2, ... 9.9, 10.

Para (2) se puede crear un array vacío del mismo tamaño que el creado en (1) y, mediante un bucle, asignarle los valores deseados en función de los valores en (1).

````python
import numpy as np
import math
x = np.linspace(0, 3*math.pi, 100) # Crea array de 100 valores
                                   # equidistantes entre 0 y 3PI
seno = np.empty(len(x))            # Crea array vacío de tamaño len(x)
coseno = np.empty(len(x))          # Crea otro array vacío de mismo tamaño
for i in range(len(x)):            # Almacena valores en seno y coseno
    seno[i] = math.sin(x[i])
    coseno[i] = math.cos(x[i])
fig, ax = plt.subplots()
ax.plot(x,seno)                   
ax.plot(x,coseno)
ax.legend(['seno', 'coseno'])
ax.grid()
None
````

## Gráficas de barras (con una única variable dependiente)

Matplotlib permite crear gráficas de muy diversos tipos. A continuación, y a modo de ejemplo, se muestra cómo crear gráficas de barras. Para ello, en lugar de *ax.plot* deberá usarse *ax.bar*

````python
x = [1,2,3,4,5]           # Valores para el eje x
y = [1,4,9,16,25]         # Valores para el eje y

fig, ax = plt.subplots() # creamos los objetos Figure y Axes

ax.bar(x,y)              # Representar los valores de y en función de x
ax.set_title("Ejemplo gráfica de barras") 
ax.set_xlabel("x")           # Etiqueta para el eje x
ax.set_ylabel("$x^2$")       # Etiqueta para el eje y
None
````

## Gráficas de barras (con más de una variable dependiente)
Si se desea representar más de un conjunto de datos mediante barras sin apilar, deberá desplazarse la x de cada conjunto (unas hacia la derecha y otras hacia la izquierda) y ajustar su anchura para que sus barras no se solapen.

````python
x = np.array([1,2,3,4,5])           # Valores para el eje x
y = np.array([1,4,9,16,25])         # Unos valores en función de x
z = np.array([1,8,27,64,125])

fig, ax = plt.subplots() # creamos los objetos Figure y Axes


ax.bar(x-0.2, y, width=0.4)     # Desplazamos las x 0.2 a la izquierda y 
                                # establecemos el ancho de barra a 0.4
ax.bar(x+0.2, z, width=0.4)     # Desplazamos las x 0.2 a la derecha y 
                                # establecemos el ancho de barra a 0.4
ax.set_title("$x^2$ vs $x^3$")  # Título
ax.set_xlabel("x")              # Etiqueta para el eje x
ax.set_ylabel("y")              # Etiqueta para el eje y
ax.legend(["$x^2$", "$x^3$"])   # Leyendas. Se proporciona una lista con una
                                # leyenda para cada una de las funciones

None
````

O, si se prefiere, pueden apilarse especicando en la llamada al método *ax.bar* el argumento *bottom* que indica las alturas y a partir de las cuales debe dibujarse cada barra:

````python
x = np.array([1,2,3,4,5]) 
y = np.array([1,4,9,16,25])        
z = np.array([1,8,27,64,125])

fig, ax = plt.subplots() # creamos los objetos Figure y Axes

ax.bar(x,y)              # Representar valores y mediante barras                                
ax.bar(x, z, bottom=y)   # Representar valores z apilados sobre los de y
ax.set_title("$x^2$ + $x^3$")         
ax.set_xlabel("x")                
ax.set_ylabel("y")                 
ax.legend(["$x^2$", "$x^3$"])                                  
None              
````

## Scatter plot o diagrama de dispersión o de puntos

El siguiente ejemplo muestra cómo crear un diagrama de puntos. Para más detalles ver la siguiente página https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

````python
import random

x,y = [],[]
for i in range(1000):
    x.append(random.uniform(0,1))
    y.append(random.uniform(0,1)**3)
fig, ax = plt.subplots()
ax.scatter(x,y, s=1, c='red')
# s permite elegir el grosor del punto
# c el color
None
````

## Varias gráficas en una misma figura

En ocasiones nos interesa crear un grid con varias gráficas en una misma figura. Para ello podemos indicarle a `plt.subplots()` el número de filas y columnas del grid como se muestra en el siguiente ejemplo. Ten en cuenta que, en ese caso, el segundo argumento devuelto ya no es un objeto de tipo `Axes` sino una matriz de objetos `Axes`:

````python
x = np.array([1,2,3,4,5])           # Valores para el eje x
y = np.array([1,4,9,16,25])         # Unos valores en función de x
z = np.array([1,8,27,64,125])

fig, ax = plt.subplots(ncols=2, figsize=(10,5)) # queremos 2 columnas

ax[0].bar(x-0.2, y, width=0.4)     # Desplazamos las x 0.2 a la izquierda y 
                                # establecemos el ancho de barra a 0.4
ax[0].bar(x+0.2, z, width=0.4)     # Desplazamos las x 0.2 a la derecha y 
                                # establecemos el ancho de barra a 0.4
ax[0].set_title("$x^2$ vs $x^3$")  # Título
ax[0].set_xlabel("x")              # Etiqueta para el eje x
ax[0].set_ylabel("y")              # Etiqueta para el eje y
ax[0].legend(["$x^2$", "$x^3$"])   # Leyendas. Se proporciona una lista con una
                                # leyenda para cada una de las funciones

x2 = np.linspace(0, 3*math.pi, 100) # Crea array de 100 valores
                                   # equidistantes entre 0 y 3PI
seno = np.empty(len(x2))            # Crea array vacío de tamaño len(x)
coseno = np.empty(len(x2))          # Crea otro array vacío de mismo tamaño
for i in range(len(x2)):            # Almacena valores en seno y coseno
    seno[i] = math.sin(x2[i])
    coseno[i] = math.cos(x2[i])
ax[1].plot(x2,seno)                   
ax[1].plot(x2,coseno)
ax[1].legend(['seno', 'coseno'])
ax[1].grid()    
      
None
````

Si en lugar de 1 dimensión damos 2 hay que acceder con 2 índices a la matriz `ax`:

````python
x = np.array([1,2,3,4,5])           # Valores para el eje x
y = np.array([1,4,9,16,25])         # Unos valores en función de x
z = np.array([1,8,27,64,125])

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5)) # grid 2x2

ax[0,0].bar(x-0.2, y, width=0.4)     # Desplazamos las x 0.2 a la izquierda y 
                                # establecemos el ancho de barra a 0.4
ax[0,0].bar(x+0.2, z, width=0.4)     # Desplazamos las x 0.2 a la derecha y 
                                # establecemos el ancho de barra a 0.4
ax[0,0].set_title("$x^2$ vs $x^3$")  # Título
ax[0,0].set_xlabel("x")              # Etiqueta para el eje x
ax[0,0].set_ylabel("y")              # Etiqueta para el eje y
ax[0,0].legend(["$x^2$", "$x^3$"])   # Leyendas. Se proporciona una lista con una
                                # leyenda para cada una de las funciones

ax[0,1].plot(x,y)    
ax[0,1].plot(x,z)
    
    
x2 = np.linspace(0, 3*math.pi, 100) # Crea array de 100 valores
                                   # equidistantes entre 0 y 3PI
seno = np.empty(len(x2))            # Crea array vacío de tamaño len(x)
coseno = np.empty(len(x2))          # Crea otro array vacío de mismo tamaño
for i in range(len(x2)):            # Almacena valores en seno y coseno
    seno[i] = math.sin(x2[i])
    coseno[i] = math.cos(x2[i])
    
ax[1,0].plot(x2,seno, label='seno')    
ax[1,0].legend()
    
ax[1,1].plot(x2,seno)                   
ax[1,1].plot(x2,coseno)
ax[1,1].legend(['seno', 'coseno'])
ax[1,1].grid()    
      
None
````
