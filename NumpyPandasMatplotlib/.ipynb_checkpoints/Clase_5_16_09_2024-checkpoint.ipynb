1. Mostrar por pantalla los datos del pasajero con identificador 148.
1. Mostrar por pantalla las filas pares del DataFrame.
1. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.
1. Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.
1. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
1. Eliminar del DataFrame los pasajeros con edad desconocida.
1. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.
1. Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.
1. Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.

```python
# Importación de librerias
import pandas as pd
import numpy as np
```

```python
# 0. Generar un DataFrame con los datos del fichero
titanic = pd.read_csv("https://aprendeconalf.es/docencia/python/ejercicios/soluciones/pandas/titanic.csv")
titanic.head(2)
```

```python
# 1. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, 
# los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas

print("Sacar las dimensiones", titanic.shape, end="\n\n")
```

```python
print("Sacar el número de datos que contiene", titanic.size, end="\n\n")
```

```python
print("Sacar los nombres de las columnas, las filas y los tipos de datos de las columnas", titanic.info(), end="\n\n")
```

```python
titanic.head(10)
```

```python
titanic.tail(10)
```

```python
# 2. Mostrar por pantalla los datos del pasajero con identificador 148
titanic[titanic['PassengerId'] == 148]
```

```python
# 3. Mostrar por pantalla las filas pares del DataFrame.

# Primera manera:
titanic.iloc[::2]

# Segunda manera:
titanic[titanic.index % 2 == 0]

# Tercera manera
titanic[titanic.index.to_series().apply(lambda x: x % 2 == 0)]
```

```python
# 4. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.
titanic[titanic['Pclass'] == 1]['Name'].sort_values()
```

```python
# 5. Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron
# Personas que murieron 
titanic['Survived'].value_counts() * 100
```

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
# 5. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase
# se puede hacer por groupby y pivottable
pd.pivot_table(titanic, index='Pclass', values='Survived', aggfunc='mean') * 100
```

```python
pd.groupby(by="Pclass")[['Survived']].mean()*100
```

```python
titanic.groupby(by="Pclass")[['Survived']].value_counts(normalize=True)
```

```python
# Que cuantos hombres y mujeres de primera clase viajaban
titanic.groupby(by=['Pclass'])[['Sex']].value_counts()
```

```python
titanic.groupby(by=['Pclass'])
```

```python
titanic.groupby(by=['Pclass']).Sex.groups
```

```python
bicis.averiadas.value_counts().sort_index()
```

```python
# Teoria 399: Dummy
# get_dummies() convierte una columna categórica en varias columnas binarias (0 o 1), una por cada categoría posible.
# Así, en lugar de tener una sola columna con valores categóricos, tendrás varias columnas con indicadores binarios.

data = {
    'a':[1,2,3],
    'b':['a', 'b', 'c'],
    'c': ['x', 'y', 'y']
}

df=pd.DataFrame(data)
df
```

```python
pd.get_dummies(df)
```

```python
df = pd.DataFrame({'color': ['rojo', 'verde', 'azul', 'rojo']})
df_dummies = pd.get_dummies(df['color'])
print(df_dummies)
```
