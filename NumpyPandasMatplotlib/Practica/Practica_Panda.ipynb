```python
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1,2,3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value1': [4,5,6]
})

# Devolver las filas que tienen claves que coinciden para ambos dataframes
inner_merged = pd.merge(df1, df2, on='key', how='inner')
print(inner_merged)

# Se puede hacer lo mismo pero con join, La diferencia de estos es que estos dataframes deben de tener indices coincidentes.
# Es decir, solo si los indices coinciden se mostraran
df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

print(df5.join(df6, how='inner'))

# Devolver la mezcla de los dos dataframes, los que no coinciden en el otro lado se rellenan con Nans
outer_merged = pd.merge(df1, df2, on='key', how="outer")
print(outer_merged)

# Merge a la izquierda y a la derechaabs
left_merged = pd.merge(df1, df2, on='key', how="left")
print(left_merged)

right_merged = pd.merge(df1, df2, on='key', how="right")
print(right_merged)
```

```python
# Al trabajar con pandas  es crucial que las columnas que trabajen con fechas y horas estén en el formato adecuado
# Cuando extraemos las fechas de ficheros es normal que se representen como strings

datos = pd.read_csv('./online_retail.csv')

# Primer paso: Imprimimos los datos para ahcernos una idea:
print(datos.head(5), end='\n\n')

# Segundo paso, vamos a ver la información para ver el tipo de las columnas
print(datos.info(),end='\n\n')
# Vemos que el tipo de datos de las columnas son object,int64,float.... 

# Esto lo que hace es convertir las fechas en tipo fecha
datos['InvoiceDate'] = pd.to_datetime(datos['InvoiceDate'])

# Ahora podemos volver a imprimir las columnas y la info:
print(datos.head(5), end='\n\n')
print(datos.info(),end='\n\n')
```
