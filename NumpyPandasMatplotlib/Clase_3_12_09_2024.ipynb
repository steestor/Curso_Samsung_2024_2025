```python
import pandas as pd
import numpy as np
```

```python
# Borrar filas y columnas de panda

df = pd.DataFrame({
    "math": [90,80,70],
    "eng": [98,89,95],
    "music": [85,95,100],
    "science": [100,90,90],
})

df.index = ["stu1", "stu2", "stu3"]

df.drop("stu2", axis=0)
```

```python
df.drop("stu2")
```

```python
# axis=0 o axis='index': Indica que deseas eliminar filas.
# axis=1 o axis='columns': Indica que deseas eliminar columnas.
```

```python
# loc: Selecciona datos por nombre (etiqueta de filas o columnas).
# iloc: Selecciona datos por posición (índices numéricos de filas o columnas).
```

```python
print(df.loc["stu1"], end="\n\n") # se pueden localizar filas solo o filas y columnas, pero NO se pueden alterar el orden
print(df.loc["stu1","eng"], end="\n\n")
print(df.loc["stu1",["math", "eng"]])  # df.loc[FILA o [FILAX,FILAY,....] , COLUMNA o [COLUMNAX, COLUMNAY, ...]]
print(df.loc["stu1","math": "eng"]) # dl.loc[FILA, COLUMNAX:COLUMNAY]  se le pueden especificar rangos con slicing


print(df.iloc[0], end="\n\n")
print(df.iloc[0,0], end="\n\n")
```

```python
# Cuando queremos asignar nuevas filas o columnas a los dataframes: df.loc[3] = 0

df.loc[3] =0 
df
```

```python
df.loc[3] = [1,2,3,4]
df
```

```python
# si quisieramos resetar todos los indices usaremos la siguiente instrucción, RECUERDA: si quieres guardar esto reasigna
df.reset_index()
```

```python
# Teoria de conjuntos

leftBlue = pd.DataFrame({
    "Name": ["Harry Potter", "David Baker", "John Smith", "Juan Martinez", "Jane Connor", np.nan, np.nan],
    "Gender": ["Male","Male","Male","Male", "Female", np.nan, np.nan],
    "Age": [23,31,22,36,30, np.nan, np.nan],
})

rightGreen = pd.DataFrame({
    "Name": [np.nan, np.nan, "John Smith", np.nan, "Janer Connor", "Alex Du Bois", "Joanne Rowling"],
    "Position": [np.nan, np.nan, "Intern", np.nan, "Manager", "Team Lead", "Manager"],
    "Wage": [np.nan, np.nan, 25000, np.nan, 70000, 75000, 90000],
})
```

```python
pd.merge(leftBlue,rightGreen, on="Name")
```

```python
# Lo que tienen en comun ambos cojuntos
pd.merge(leftBlue,rightGreen, left_on="Name", right_on="Name", how="inner") 
```

```python
# Lo que tienen en común los conjuntos más lo que tienen a la izquierda
pd.merge(leftBlue,rightGreen, left_on="Name", right_on="Name", how="left")
```

```python
# Lo que tienen en común más lo de la derecha
pd.merge(leftBlue,rightGreen, left_on="Name", right_on="Name", how="right")
```

```python
# Las final que no tienen coincidencia
pd.merge(leftBlue,rightGreen, left_on="Name", right_on="Name", how="outer")
```
