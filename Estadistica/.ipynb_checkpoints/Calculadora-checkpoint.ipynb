# CORRELACIÓN

La correlación es una medida que indica la fuerza y dirección de una relación lineal entre dos variables. Su valor siempre está entre -1 y 1.

Una correlación de 1 indica una relación lineal positiva perfecta.
Una correlación de -1 indica una relación lineal negativa perfecta.
Una correlación de 0 indica que no hay relación lineal.



Pero las variables PUEDEN ser DEPENDIENTES

Calculo de la correlacion:




```python
def calculate_correlation(values_X, values_Y, probability):
    # Calcular E[XY]
    E_XY = sum([x * y * probability for x, y in zip(values_X, values_Y)])
    
    # Calcular E[X] y E[Y]
    E_X = sum([x * probability for x in values_X])
    E_Y = sum([y * probability for y in values_Y])
    
    # Fórmula de la correlación
    correlation = E_XY - (E_X * E_Y)
    
    return correlation
```

## Ejercicio diapositiva  108

Esta ruleta representa nuestra variable X. Cada vez que la giramos, tenemos la misma probabilidad (1/3) de que caiga en -1, 0 o 1:



Y sabemos que:


Por ejemplo:

Si sale -1, entonces X = -1 y Y = (-1)² = 1
Si sale 0, entonces X = 0 y Y = 0² = 0
Si sale 1, entonces X = 1 y Y = 1² = 1

Observa que Y solo puede ser 0 o 1, nunca negativo.
Lo que el ejercicio nos está pidiendo es que calculemos si hay alguna relación entre X e Y.


```python
# Primer paso: Calculamos P(X)
E_x = (-1*1/3) + (0*1/3) +(1*1/3)
print("E_x", E_x)

# Segundo paso: Calculamos P(Y)
# Y=0 cuando X=0
# Y=1 cuando X=1 o X=-1
E_y = (0*1/3) + (1*2/3)
print("E_y", E_y)

# Tercer paso: calculamos E[XY]
# XY = X^3 porque Y=X^2
# Multiplicamos por la probabilidad correspondiente y sumamos
E_XY = ((-1**3) * (1/3)) + ((0**3) * (1/3)) + ((1**3) * (1/3))
print("E_XY",E_XY)

# Calcular la correlación
calculate_correlation([-1,0,1], [-1,0,1], E_XY)
# O lo que es lo mismo: 
# Cor(X,Y) = E[XY] - E[X] × E[Y]
# E[XY] = 0, E[X] = 0, E[Y] = 2/3
# 0 - 0 × (2/3) = 0

# El resultado final de 0 indica que no hay una relación lineal entre X e Y.

# Ahora vamos a averiguar si son dependientes o independientes las variables
```

Para saber si son independientes deberemos validar esto: 
P(X,Y) == P(X) * P(Y) 



Si se cumple que P(X,Y) = P(X) * P(Y) para todos los valores posibles de X e Y, entonces las variables son independientes. 

En nuestro caso específico (X e Y = X²):

Cuando X aumenta de -1 a 0, Y disminuye de 1 a 0.
Cuando X aumenta de 0 a 1, Y aumenta de 0 a 1.
Este comportamiento "mixto" (a veces aumenta, a veces disminuye) resulta en una correlación neta de 0.



Como podemos ver no podemos trazar una línea recta.

En resumen, una correlación de 0 nos dice que no hay una tendencia lineal consistente entre X e Y, pero no excluye otros tipos de relaciones más complejas, como la relación cuadrática que tenemos en este caso.
