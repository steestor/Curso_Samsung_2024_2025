```python
#SOLUCIÓN
ejemplo1 = pd.read_excel('./EjemploAjuste curvas.xlsx')
df = ejemplo1[ejemplo1.columns[:2]]
df

# Separamos los datos en entrenamiento y test
train_data, test_data = train_test_split(df, test_size=0.5, random_state=42)

print(f"Estos son los datos de entrenamiento {len(train_data)}")
print(f"Estos son los datos de entrenamiento {len(test_data)}")

# Extraer X y y para el modelo lineal
X = train_data.iloc[:, 0].values  # Valores de la primera columna
y = train_data.iloc[:, 1].values  # Valores de la segunda columna

# Añadir un vector columna de unos para el término de intersección (B0)
X = np.vstack([X, np.ones(len(X))]).T

# Calcular los coeficientes usando la fórmula de mínimos cuadrados
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# Mostrar los coeficientes B0 y B1
print(f"Coeficiente B1: {beta[0]}")
print(f"Intercepto B0: {beta[1]}")

# Generar predicciones usando el modelo lineal
y_pred = X @ beta

# Sacamos las predicciones de 
print(y_pred)

# Sacamos el erro cuadratico
print((y - y_pred).T @ (y - y_pred))
```
