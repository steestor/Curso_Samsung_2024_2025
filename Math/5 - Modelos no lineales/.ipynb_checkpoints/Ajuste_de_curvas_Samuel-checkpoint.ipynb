```python
def calculoError(x, y):
    return np.mean((x - y)**2)

def calculoErrorCuadratico(a, b, x, y, totaldata):
    y_hat = a * x + b
    error = np.sum((y - y_hat) ** 2)
    return error/np.round((totaldata/2))
 
def iteracion_valores_alpha(a_values, b_values , x, y,totaldata):
    results = []
    for a in a_values:
        for b in b_values:
            error = calculoErrorCuadratico(a,b,x,y,totaldata)
            results.append((a, b, error))

            print(f"a = {a}, b = {b}, Error cuadratico = {error}")
        print()
            
def dividir_datos(datos, proporcion=0.5):
    datos = np.array(datos)
    
    # Obtener el número total de elementos
    n = len(datos)
    
    # Crear un array de índices y mezclarlo
    indices = np.arange(n)
    np.random.seed(40)
    np.random.shuffle(indices)
    
    # Calcular el punto de división
    punto_division = int(n * proporcion)
    
    # Dividir los datos
    parte1 = datos[indices[:punto_division]]
    parte2 = datos[indices[punto_division:]]
    
    return parte1, parte2
```

```python
x = datos["X"].dropna().to_numpy()
y = datos["Y"].dropna().to_numpy()

x_ent, x_pre = dividir_datos(x)
y_ent, y_pre = dividir_datos(y)

a_values = [0.1, 0.2, 1, 1.1, 1.2, 2, 2.5, 3, 3.5, 4, 4.5, 5 ,5.5, 20, 40, 45]
b_values = [0.1, 0.2, 1, 1.1, 1.2, 2, 2.5, 3, 3.5, 4, 4.5, 5 ,5.5,20, 40, 45]

iteracion_valores_alpha(a_values, b_values, x_ent, y_ent,15)
```

# ¿Cuando paramos?
Deberíamos parar cuando encontremos el error cuadrático mínimo global. En este caso, basándonos en los resultados mostrados, el error cuadrático mínimo se alcanza cuando a = 2.5 y b = 2.5, con un error de 1091.0096447265626. Sin embargo, es importante notar que este podría ser solo un mínimo local.

# ¿Y si vamos en circulo?
Si vamos en círculos, significa que nuestro método de búsqueda está oscilando alrededor de un punto y no está convergiendo a un mínimo. En este caso, podríamos considerar:

Reducir el tamaño del paso en nuestros valores de a y b.
Implementar un criterio de parada basado en la mejora mínima del error entre iteraciones.
Utilizar un método de optimización más avanzado, como el descenso de gradiente con momento.

#¿Y si no es min global?
Si el mínimo encontrado no es global, significa que hemos encontrado un mínimo local. Para abordar esto, podríamos:

Ampliar el rango de valores de a y b para explorar más el espacio de parámetros.
Utilizar técnicas de optimización global, como algoritmos genéticos o simulated annealing.
Iniciar la búsqueda desde múltiples puntos de partida aleatorios.
Implementar un método de búsqueda en cuadrícula más fino alrededor del mejor punto encontrado.
