```python
import pandas as pd
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
from PIL import Image
```

```python
A = np.array([[1,0,1],[0,1,1],[0,0,2]])
print(A)
```

```python
# Autovalor y autovector
autoval, autovec = la.eig(A)
print("autoval:")
print(autoval) 
print("autovec: ") 
print(autovec)
```

```python
la.det(autovec) # Los tres vectores propios son linenalmente independientes
```
