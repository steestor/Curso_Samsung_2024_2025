# Modelos lineales de varias variables

Vamos a predecir la temperatura mediante lo siguiente: 
t=a*p + b*h + c
    - donde t será la temperatura
    - donde p será la presión
    - donde h será la humedad

Noramlizaremos los datos de entrada mediante la siguiente formula:
x = (x - la media de x)/la desviacion estándar de x

Pasos a seguir:

1 - Normalizamos

2 - Dividimos en partes iguales los datos de entrenamiento y los datos de validación

3 - Ajustamos los modelos

4 - Comprobamos que los datos no están sobre entrenados

5 - Interpretamos los resultados

6 - Comparamos los resultados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("./DataTemperaturePressureHumidity.xlsx")
data.head()
```

```python
# 1. Normalizar los datos
def normalize(x):
    return (x - x.mean()) / x.std()

data['Humidity_norm'] = normalize(data['Humidity'])
data['Pressure_norm'] = normalize(data['Pressure'])
data['Temperature_norm'] = normalize(data['Temperature'])

data.head()
```

```python
# 2. Dividir los datos en conjuntos de entrenamiento y validación con una proporción aproximada de 50/50
np.random.seed(42)
msk = np.random.rand(len(data)) < 0.5
train = data[msk]
test = data[~msk] # el simbolo ~ invierte la máscara booleana 
```

```python
train.head()
```

```python
test.head()
```

```python
# 3. Ajusta un modelo de regresión lineal utilizando la fórmula de mínimos cuadrados ordinarios.
# La fórmula de mínimos cuadrados ordinarios  es un método fundamental en estadística para estimar los parámetros de un modelo de regresión lineal. 
# La formula es: β̂ = (X^T X)^(-1) X^T y
def fit_linear_regression(X, y):
    X = np.column_stack((np.ones(X.shape[0]), X))
    coeffs = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return coeffs

X_train = train[['Pressure_norm', 'Humidity_norm']]
y_train = train['Temperature_norm']
coeffs = fit_linear_regression(X_train, y_train)
```

```python
coeffs
```

```python
# 4. Comprobar el sobreajuste. Comprueba si hay sobreajuste comparando el rendimiento en los conjuntos de entrenamiento y prueba, 
# utilizando funciones personalizadas para calcular el Error Cuadrático Medio (MSE) y Coeficiente de Determinación (R²)
def predict(X, coeffs):
    X = np.column_stack((np.ones(X.shape[0]), X))
    return X.dot(coeffs)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def r2_score(y_true, y_pred):
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    ss_res = np.sum((y_true - y_pred)**2)
    return 1 - (ss_res / ss_tot)

# Nueva función para calcular el error cuadrático total
def error_cuadratico_total(y_true, y_pred):
    return np.sum((y_true - y_pred)**2)

# Cálculo de predicciones
train_pred = predict(X_train, coeffs)
test_pred = predict(test[['Pressure_norm', 'Humidity_norm']], coeffs)

# Cálculo de métricas
train_mse = mse(y_train, train_pred)
test_mse = mse(test['Temperature_norm'], test_pred)
train_r2 = r2_score(y_train, train_pred)
test_r2 = r2_score(test['Temperature_norm'], test_pred)

# Cálculo del error cuadrático total
train_sse = error_cuadratico_total(y_train, train_pred)
test_sse = error_cuadratico_total(test['Temperature_norm'], test_pred)

# Impresión de resultados
print(f"Train MSE (Error Cuadrático Medio): {train_mse:.4f}")
print(f"Test MSE (Error Cuadrático Medio): {test_mse:.4f}")
print(f"Train R2 (Coeficiente de Determinación): {train_r2:.4f}")
print(f"Test R2 (Coeficiente de Determinación): {test_r2:.4f}")
print(f"Train SSE (Error Cuadrático Total): {train_sse:.4f}")
print(f"Test SSE (Error Cuadrático Total): {test_sse:.4f}")

# Si el MSE de entrenamiento es mucho menor que el MSE de prueba, esto puede indicar sobreajuste.
# Si el R² de entrenamiento es mucho mayor que el R² de prueba, esto también puede indicar sobreajuste.
# Idealmente, queremos que estos valores sean similares para los conjuntos de entrenamiento y prueba.

# Ejemplo:
# Train MSE: 0.2000
# Test MSE: 0.8000
# Train R2: 0.9000
# Test R2: 0.5000
# Esto sugeriría un fuerte sobreajuste, ya que el modelo funciona mucho mejor en los datos de entrenamiento que en los de prueba.

# Pero si vemos lo siguiente:
# Train MSE: 0.3000
# Test MSE: 0.3200
# Train R2: 0.7500
# Test R2: 0.7300
# Esto sugeriría un buen equilibrio, ya que el rendimiento es similar en ambos conjuntos de datos.a

# Esta comparación nos ayuda a evaluar si nuestro modelo es generalizable a nuevos datos o si está demasiado ajustado a los
# datos de entrenamiento específicos.
```

```python
# 5. Interpreta los resultados mostrando la ecuación del modelo
c, a, b = coeffs

print(f"\nModelo: Temperatura = {a:.4f} * Presión + {b:.4f} * Humedad + {c:.4f}")
```

```python
import numpy as np
import pandas as pd

# Crear datos de ejemplo
np.random.seed(42)
n_samples = 1000
df = pd.read_excel("./DataTemperaturePressureHumidity.xlsx")

# Normalización
def normalize(data):
    return (data - np.mean(data)) / np.std(data)

df_norm = df.apply(normalize)

# División de datos
np.random.seed(42)
msk = np.random.rand(len(df_norm)) < 0.5
train = df_norm[msk]
test = df_norm[~msk]

# Definición de modelos
def model1(X):
    return np.column_stack([np.sqrt(np.abs(X['Humidity'])), X['Pressure'], np.ones(X.shape[0])])

def model2(X):
    return np.column_stack([X['Humidity'], np.ones(X.shape[0])])

def model3(X):
    return np.column_stack([X['Pressure'], np.ones(X.shape[0])])

models = [
    ("t = a*√|h| + b*p + c", model1),
    ("t = a*h + b", model2),
    ("t = a*p + b", model3)
]

# Función robusta para ajustar modelo y calcular métricas
def fit_and_evaluate_robust(X, y, X_test, y_test):
    # Asegurarnos de que y sea un vector columna
    y = y.reshape(-1, 1)
    y_test = y_test.reshape(-1, 1)
    
    # Usar SVD para una solución más estable
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    s_inv = np.where(s > 1e-15, 1/s, 0)
    coeffs = Vt.T @ (s_inv[:, np.newaxis] * (U.T @ y))
    
    y_pred = X @ coeffs
    y_test_pred = X_test @ coeffs
    
    train_mse = np.mean((y - y_pred)**2)
    test_mse = np.mean((y_test - y_test_pred)**2)
    
    y_mean = np.mean(y)
    ss_tot = np.sum((y - y_mean)**2)
    ss_res = np.sum((y - y_pred)**2)
    train_r2 = 1 - (ss_res / ss_tot)
    
    y_test_mean = np.mean(y_test)
    ss_tot_test = np.sum((y_test - y_test_mean)**2)
    ss_res_test = np.sum((y_test - y_test_pred)**2)
    test_r2 = 1 - (ss_res_test / ss_tot_test)
    
    return coeffs, train_mse, test_mse, train_r2, test_r2

# Ajuste de modelos y evaluación
results = []

for name, model_func in models:
    X_train = model_func(train)
    X_test = model_func(test)
    
    try:
        coeffs, train_mse, test_mse, train_r2, test_r2 = fit_and_evaluate_robust(
            X_train, train['Temperature'].values, X_test, test['Temperature'].values
        )
        
        results.append({
            'Model': name,
            'Coefficients': coeffs.flatten(),  # Aplanar los coeficientes para facilitar la impresión
            'Train MSE': train_mse,
            'Test MSE': test_mse,
            'Train R2': train_r2,
            'Test R2': test_r2
        })
    except Exception as e:
        print(f"Error en el modelo {name}: {e}")
        continue

# Análisis de resultados
for result in results:
    print(f"\nModelo: {result['Model']}")
    print(f"Coeficientes: {result['Coefficients']}")
    print(f"Train MSE: {result['Train MSE']:.4f}")
    print(f"Test MSE: {result['Test MSE']:.4f}")
    print(f"Train R2: {result['Train R2']:.4f}")
    print(f"Test R2: {result['Test R2']:.4f}")
    
    mse_diff = (result['Test MSE'] - result['Train MSE']) / result['Train MSE'] * 100
    r2_diff = (result['Train R2'] - result['Test R2']) / result['Train R2'] * 100
    
    print(f"Diferencia porcentual en MSE (Test - Train): {mse_diff:.2f}%")
    print(f"Diferencia porcentual en R2 (Train - Test): {r2_diff:.2f}%")
    
    if mse_diff > 10 or r2_diff > 10:
        print("Posible sobreentrenamiento")
    else:
        print("No hay evidencia clara de sobreentrenamiento")

# Identificación del mejor modelo
if results:
    best_model = max(results, key=lambda x: x['Test R2'])
    print(f"\nEl mejor modelo basado en R2 de test es: {best_model['Model']}")
    print(f"R2 de test: {best_model['Test R2']:.4f}")
    print(f"MSE de test: {best_model['Test MSE']:.4f}")
else:
    print("\nNo se pudo ajustar ningún modelo correctamente.")
```
