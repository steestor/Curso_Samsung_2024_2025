```python
data = pd.read_excel("./CurrencyExhangeRatesSimplificado.xlsx")
data
```

```python
data = data[["Chinese Yuan", "Russian Ruble", "U.K. Pound Sterling", "Euro"]]
X = data[["Chinese Yuan", "Russian Ruble", "U.K. Pound Sterling"]] # Usar Yuan, Ruble, Pound del día anterior
y = data["Euro"]

data = data.to_numpy()
data
```

```python
# Función para realizar la regresión lineal por mínimos cuadrados
def linear_regression(X, y):
    X = np.column_stack([np.ones(X.shape[0]), X])
    coeffs = np.linalg.inv(X.T @ X) @ X.T @ y
    return coeffs[0], coeffs[1:]

# Función para hacer predicciones
def predict(X, intercept, coeffs):
    return intercept + X @ coeffs

# Función para calcular el error cuadrático medio (MSE)
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# EJERCICIO 1: Ajusta un modelo lineal para predecir los valores de EUR en un día a partir de los valores del día anterior de
# EUR, US Dollar, UK Pound, Yuan y CAD Dollar

# Dividir los datos en entrenamiento y validación (50% cada uno)
split = len(X) // 2
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]

# Ajustar el modelo con los datos de entrenamiento
intercept, coeffs = linear_regression(X_train, y_train)

# Hacer predicciones en los conjuntos de entrenamiento y validación
y_train_pred = predict(X_train, intercept, coeffs)
y_val_pred = predict(X_val, intercept, coeffs)

# Calcular el MSE para entrenamiento y validación
mse_train = mse(y_train, y_train_pred)
mse_val = mse(y_val, y_val_pred)

# Imprimir resultados
print("Modelo lineal para predecir EUR:")
print(f"Intercept: {intercept}")
print(f"Coeficientes: Yuan: {coeffs[0]}, Ruble: {coeffs[1]}, Pound: {coeffs[2]}")
print(f"MSE (Entrenamiento): {mse_train}")
print(f"MSE (Validación): {mse_val}")

# La ecuación del modelo es:
print(f"Ecuación del modelo: EUR = {intercept:.4f} + {coeffs[0]:.4f}*Yuan + {coeffs[1]:.4f}*Ruble + {coeffs[2]:.4f}*Pound")
```

### Rendimiento del modelo:
Buen ajuste a los datos de entrenamiento (MSE bajo).
Cierto grado de sobreajuste, pero aún mantiene capacidad predictiva en datos de validación.
    
### Influencia de las monedas:
La Libra Esterlina tiene el mayor impacto positivo en el Euro.
El Yuan muestra una ligera relación inversa con el Euro.
El Rublo tiene una influencia mínima en este modelo.
    
### Limitaciones:
El modelo asume relaciones lineales, lo cual puede ser una simplificación excesiva para mercados financieros.
No considera otros factores económicos o geopolíticos relevantes.

### Aplicación práctica:
Útil como referencia, pero no debe ser la única herramienta para toma de decisiones financieras.
Se recomienda usar con cautela debido al sobreajuste observado.
    
###  Posibles mejoras:
Considerar incluir más variables o utilizar modelos más complejos para capturar mejor la dinámica del mercado de divisas.

```python
# EJERCICIO 2: Ajusta un modelo lineal que utilice información de los dos días previos

def linear_regression(X, y):
    X = np.column_stack([np.ones(X.shape[0]), X])
    coeffs = np.linalg.inv(X.T @ X) @ X.T @ y
    return coeffs[0], coeffs[1:]

def predict(X, intercept, coeffs):
    return intercept + X @ coeffs

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Preparar X e y para el modelo de dos días previos
X = np.column_stack([data[:-2, :3], data[1:-1, :3]])  # Yuan, Ruble, Pound de los dos días anteriores
y = data[2:, 3]  # EUR del día actual

# Dividir los datos en entrenamiento y validación (50% cada uno)
split = len(X) // 2
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]

# Ajustar el modelo con los datos de entrenamiento
intercept, coeffs = linear_regression(X_train, y_train)

# Hacer predicciones en los conjuntos de entrenamiento y validación
y_train_pred = predict(X_train, intercept, coeffs)
y_val_pred = predict(X_val, intercept, coeffs)

# Calcular el MSE para entrenamiento y validación
mse_train = mse(y_train, y_train_pred)
mse_val = mse(y_val, y_val_pred)

# Imprimir resultados
print("Modelo lineal para predecir EUR usando datos de dos días previos:")
print(f"Intercept: {intercept}")
print("Coeficientes:")
for i, coef in enumerate(coeffs):
    day = "t-1" if i < 3 else "t-2"
    currency = ["Yuan", "Ruble", "Pound"][i % 3]
    print(f"{currency} ({day}): {coef}")
print(f"MSE (Entrenamiento): {mse_train}")
print(f"MSE (Validación): {mse_val}")

# La ecuación del modelo es más compleja ahora, así que la presentaremos de forma más legible
print("\nEcuación del modelo:")
print(f"EUR = {intercept:.4f}")
for i, coef in enumerate(coeffs):
    day = "t-1" if i < 3 else "t-2"
    currency = ["Yuan", "Ruble", "Pound"][i % 3]
    print(f"  + {coef:.4f} * {currency}({day})")
```
